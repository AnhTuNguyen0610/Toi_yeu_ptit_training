**YÊU CẦU 1.1: Học viên tạo một ma trận vuông cấp n, bao gồm các phần tử, là những số nguyên ngẫu nhiên trong khoảng [0-100]**

````python
import numpy as np
````

````python
#{HỌC VIÊN VIẾT CODE VÀO ĐÂY}

n = int(input("Nhập cấp của ma trận: "))

matrix = np.random.randint(0, 101, size=(n, n))

print(matrix)




````

**YÊU CẦU 1.2: Sử dụng ma trận tạo được trong yêu cầu 1, Học viên tạo 2 vector như sau:**

**v_chinh:** bao gồm các phần tử nằm trên đường chéo chính của ma trận.

**V_phu:** bao gồm các phần tử nằm trên đường chéo phụ của ma trận

````python
v_chinh = np.diag(matrix)

v_phu = np.diag(np.fliplr(matrix))

print("Vector đường chéo chính :")
print(v_chinh)

print("Vector đường chéo phụ :")
print(v_phu)
````

**YÊU CẦU 1.3: Nhập vào số nguyên x bất kỳ trong khoảng (0-100), đếm xem có bao nhiêu phần tử trong ma trận sinh ra ở yêu cầu 1 có giá trị bằng x, lớn hơn x, nhỏ hơn x và hiển thị kết quả:**

````python
#{HỌC VIÊN VIẾT CODE VÀO ĐÂY}

x=int(input("Nhap so nguyen x: "))

lon=np.sum(matrix > x)
bang=np.sum(matrix == x)
nho=np.sum(matrix < x)

print(f"Số phần tử bằng {x}: {bang}")
print(f"Số phần tử lớn hơn {x}: {lon}")
print(f"Số phần tử nhỏ hơn {x}: {nho}")

````

**YÊU CẦU 2.1: Sử dụng dữ liệu bảng điểm của lớp 2A. Cho biết:**

2. Điểm TB cao nhất của học sinh trong lớp.
3. Điểm trung bình thấp nhất của học sinh trong lớp

````python
path = "/content/drive/MyDrive/for_student/Bai4/ThucHanh/Data_Excercise/Diem_2A.txt"
````

````python
data = np.loadtxt(path, delimiter=',')

dtb_hocsinh = np.mean(data, axis=1)

max_avg = np.max(dtb_hocsinh)
hs_max = np.where(dtb_hocsinh == max_avg)[0] + 1  # +1 để ra số thứ tự học sinh (1-based)

min_avg = np.min(dtb_hocsinh)
hs_min = np.where(dtb_hocsinh == min_avg)[0] + 1

print("ĐTB của từng học sinh")
for i, dtb in enumerate(dtb_hocsinh, start=1):
    print(f"Học sinh {i:2d}: {dtb:.2f}")

print("\nĐiểm TB cao nhất")
for i in hs_max:
    print(f"Học sinh {i}: {max_avg:.2f}")

print("\nĐiểm trung bình thấp nhất")
for i in hs_min:
    print(f"Học sinh {i}: {min_avg:.2f}")
````

**YÊU CẦU 2.2: Sử dụng dữ liệu bảng điểm của lớp 2A. Cho biết:**

1. ĐTB của từng môn học.
2. Điểm TB cao nhất trong các môn học là bao nhiêu?
3. Điểm TB thấp nhât trong các môn hoc là bao nhiêu?

````python
#{HỌC VIÊN VIẾT CODE VÀO ĐÂY}

dtb_monhoc = np.mean(data, axis=0)

max_avg = np.max(dtb_monhoc)
hs_max = np.where(dtb_monhoc == max_avg)[0] + 1  # +1 để ra số thứ tự học sinh (1-based)

min_avg = np.min(dtb_monhoc)
hs_min = np.where(dtb_monhoc == min_avg)[0] + 1

print("ĐTB của từng môn học")
for i, dtb in enumerate(dtb_monhoc, start=1):
    print(f"môn học {i:2d}: {dtb:.2f}")

print("\nĐiểm TB cao nhất")
for i in hs_max:
    print(f"môn học {i}: {max_avg:.2f}")

print("\nĐiểm trung bình thấp nhất")
for i in hs_min:
    print(f"môn học {i}: {min_avg:.2f}")



````

#### **YÊU CẦU 2.3: Sử dụng dữ liệu bảng điểm của lớp 2A. Cho biết:**
---
1. Sinh viên có điểm đồng đều nhất tất cả các môn - Hiển thị bảng điểm của SV này. Sinh viên có điểm các môn lệch nhất trong lớp - Hiển thị bảng điểm của sinh viên này.
2. Môn học có điểm đồng đều nhất. Môn học có điểm chênh lệch nhất.



````python
#{HỌC VIÊN VIẾT CODE VÀO ĐÂY}

std_hs = np.std(data, axis=1)

hs_min_std = np.argmin(std_hs)
hs_max_std = np.argmax(std_hs)

std_mon = np.std(data, axis=0)

mon_min_std = np.argmin(std_mon)
mon_max_std = np.argmax(std_mon)

print("Sinh viên có điểm đồng đều nhất tất cả các môn", hs_min_std + 1)
print("Bảng điểm:", data[hs_min_std])

print("\nSinh viên có điểm các môn lệch nhất trong lớp", hs_max_std + 1)
print("Bảng điểm:", data[hs_max_std])

print(f"\nMôn đồng đều nhất: Môn {mon_min_std + 1}")
print(f"Môn chênh lệch nhất: Môn {mon_max_std + 1}")



````
