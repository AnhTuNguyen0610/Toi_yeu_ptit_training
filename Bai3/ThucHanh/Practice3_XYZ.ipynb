1. Viết hàm calculator_BMI():
   - Đầu vào là 2 tham số: Chiều cao (m), cân nặng (Kg); thiết lập giá trị mặc định bằng chiều cao, cân nặng của bạn
   - Trả về 2 giá trị: chỉ số BMI và phân loại (UnderWeight, NormalWeight, OverWeight, Obese)

Biết chỉ số BMI của người đó theo công thức:

    * BMI = Weight(Kg) / Height(m) ^2
    
Dựa vào chỉ số BMI tính được đưa ra nhận xét về cơ thể của người đó theo các mức:

    - Underweight: BMI<18.5
    - NormalWeight: 18.5 <= BMI <= 24.9
    - Overweight: 25 <= BMI <=29.9
    - Obese: BMI>=30

```python
#{HỌC VIÊN VIẾT CODE VÀO ĐÂY}

def BMI(height=1.71, weight=50):
    BMI = weight / (height ** 2)

    if BMI < 18.5:
        category = "UnderWeight"
    elif 18.5 <= BMI <= 24.9:
        category = "NormalWeight"
    elif 25 <= BMI <= 29.9:
        category = "OverWeight"
    else:
        category = "Obese"

    return round(BMI, 2), category


```

```python
bmi_value, bmi_category = BMI(1.75, 68)
print("Chỉ số BMI:", bmi_value)
print("Phân loại:", bmi_category)
```

```python
def quy_doi_diem(diem):
    if diem >= 9.0:
        return "A+", 4.0
    elif diem >= 8.5:
        return "A", 3.7
    elif diem >= 8.0:
        return "B+", 3.5
    elif diem >= 7.0:
        return "B", 3.0
    elif diem >= 6.5:
        return "C+", 2.5
    elif diem >= 5.5:
        return "C", 2.0
    elif diem >= 5.0:
        return "D+", 1.5
    elif diem >= 4.0:
        return "D", 1.0
    else:
        return "F", 0.0

def cal_point(diem_he_10):
    diem_chu = []
    diem_he_4 = []

    for d in diem_he_10:
        chu, he4 = quy_doi_diem(d)
        diem_chu.append(chu)
        diem_he_4.append(he4)

    tb_he_10 = sum(diem_he_10) / len(diem_he_10)
    tb_he_4 = sum(diem_he_4) / len(diem_he_4)

    return f"diem_he_10: {diem_he_10}\ndiem_chu: {diem_chu}\ndiem_he_4: {diem_he_4}\ntb_he_10: {tb_he_10}\ntb_he_4: {tb_he_4}\n"


diem_he_10 = [9.0, 8.5, 7.8, 6.2, 5.0, 9.5, 2.3, 4.9, 7.6]
ketqua=cal_point(diem_he_10)
print(ketqua)


```

```python
def list_prime(N):
    if N < 2:
        return []

    is_prime = [True] * (N + 1)
    is_prime[0] = is_prime[1] = False

    for i in range(2, int(N**0.5) + 1):
        if is_prime[i]:
            for j in range(i * i, N + 1, i):
                is_prime[j] = False

    return [i for i, check in enumerate(is_prime) if check]


N = int(input("Nhập vào một số nguyên N: "))
p = list_prime(N)
print(f"Các số nguyên tố từ 1 tới {N} là:")
print(p)
```

7. Xây dựng hàm thực hiện chức năng sau theo phương pháp đệ quy:
Truyền vào một số tự nhiên N,
    * Nếu N là số chẵn tính tổng tất cả các số chẵn từ 0 tới N
    * Nếu N là số lẻ tỉnh tổng tất cả các số lẻ từ 1 tới N

```python
#{HỌC VIÊN VIẾT CODE VÀO ĐÂY}

def tong(N):
    if N <= 1:
        return N

    if N % 2 == 0:
        return N + tong(N - 2)
    else:
        return N + tong(N - 2)

print(tong(5))

```

## BÀI 19.1: Xây dựng lớp Person:
---
1. Gồm các Thuộc tính:
    * họ tên (name), năm sinh (year), chiều cao height), cân nặng (weight)
    * Giá trị mặc định của các thuộc tính là thông tin của bạn

2. Gồm các Phương thức:
    * Geeting(): Hiển thị thông tin của Person
    * BMI(): Tính toán theo chỉ số BMI của Person

```python
#{HỌC VIÊN VIẾT CODE VÀO ĐÂY}

class Person:
  def __init__(self,name,year,height,weight):
    self.name=name
    self.year=year
    self.height=height
    self.weight=weight

  def geeting(self):
    print(f"Ten: {self.name}\nNam sinh: {self.year}\nChieu cao: {self.height}\nCan nang: {self.weight}")

  def cal_BMI(self):
    bmi_value, bmi_category = BMI(self.height, self.weight)
    print("Chỉ số BMI:", bmi_value)
    print("Phân loại:", bmi_category)

p=Person("Nguyen Huy Hoang",2006,1.70,50)
p.geeting()
p.cal_BMI()
```

## Bài 19.2. Xây dựng lớp Vietnam, kế thừa từ lớp Person:
---
* Kế thừa tất cả các thuộc tính của lớp Person và Bổ sung thêm Thuộc tính:
Dân tộc (nation), giá trịnh mặc định là “Kinh”
* Kế thừa phương thức BMI của lớp Person và  cập nhật lại Phương thức Geeting() (Method overriding):

Ví dụ Geeting():

Tôi tên là:  Thanh Huyền - Dân tộc: Tày

Tôi sinh năm  2000 . Rất vui được làm quyen với bạn!

```python
#{HỌC VIÊN VIẾT CODE VÀO ĐÂY

class Vietnam(Person):
    def __init__(self, name, year, height, weight, nation="Kinh"):
        super().__init__(name, year, height, weight)
        self.nation = nation
    def geeting(self):
        print(f"Tôi tên là: {self.name} - Dân tộc: {self.nation}")
        print(f"Tôi sinh năm {self.year}. Rất vui được làm quen với bạn!")


nguoi1 = Vietnam("Thanh Huyền", 2000, 1.60, 50, "Tày")
nguoi1.geeting()
nguoi1.cal_BMI()



```
