```python
#{HỌC VIÊN VIẾT CODE VÀO ĐÂY}

diem_he_10 = [9.0, 8.5, 7.8, 6.2, 5.0, 9.5, 2.3, 4.9, 7.6]

def quy_doi_diem(diem):
    if diem >= 9.0:
        return "A+", 4.0
    elif diem >= 8.5:
        return "A", 3.7
    elif diem >= 8.0:
        return "B+", 3.5
    elif diem >= 7.0:
        return "B", 3.0
    elif diem >= 6.5:
        return "C+", 2.5
    elif diem >= 5.5:
        return "C", 2.0
    elif diem >= 5.0:
        return "D+", 1.5
    elif diem >= 4.0:
        return "D", 1.0
    else:
        return "F", 0.0

diem_chu = []
diem_he_4 = []

for d in diem_he_10:
    chu, he4 = quy_doi_diem(d)
    diem_chu.append(chu)
    diem_he_4.append(he4)

tb_he_10 = sum(diem_he_10) / len(diem_he_10)
tb_he_4 = sum(diem_he_4) / len(diem_he_4)

print("Điểm hệ 10:", diem_he_10)
print("Điểm chữ  :", diem_chu)
print("--------ĐIỂM TRUNG BÌNH--------")
print(f"Tổng số môn học:{len(diem_he_10)}")
print("ĐTB hệ 10:", tb_he_10)
print("ĐTB hệ 4 :", tb_he_4)





```

## Bài 11: Kiểm tra N có là số nguyên tố?
---
Nhập vào số tự nhiên N, cho biết N có phải là số nguyên tố hay không? (Số nguyên tố là số lớn hơn 1, chỉ chia hết cho 1 và chính nó)


```python
#{HỌC VIÊN VIẾT CODE VÀO ĐÂY}

def SNT(n):
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False
    return True


```

```python
# Kiểm tra với các giá trị sau
#N0 = 2, 4        --> ???
#N1 = 98711653011 --> ??
#N2 = 87178291199 --> ??
```

```python
num = list(map(int, input("Nhập các số: ").split()))

for n in num:
    if SNT(n):
        print(f"{n} là số nguyên tố.")
    else:
        print(f"{n} không phải là số nguyên tố.")
```

## Bài 12: dãy số nguyên tố
---
Nhập vào một số N, hiển thị các số nguyên tố từ 2 tới N:


```python
#{HỌC VIÊN VIẾT CODE VÀO ĐÂY}

def sang_SNT(N):
    if N < 2:
        return []

    is_prime = [True] * (N + 1)
    is_prime[0] = is_prime[1] = False

    for i in range(2, int(N**0.5) + 1):
        if is_prime[i]:
            for j in range(i * i, N + 1, i):
                is_prime[j] = False

    return [i for i, check in enumerate(is_prime) if check]





```

```python
N = int(input("Nhập vào một số nguyên N: "))
p = sang_SNT(N)
print(f"Các số nguyên tố từ 1 tới {N} là:")
print(p)
```

Khởi tạo một biến tuple lưu chiều cao của sinh viên trong một lớp (m). Thực hiện:

    * 1) Hiển thị chiều cao của sinh viên cao nhất – thấp nhất trong lớp
    * 2) Tính chiều cao trung bình (TB) của sinh viên trong lớp.
    * 3) Số lượng sinh viên trong lớp có chiều cao bằng chiều cao TB của lớp.
    * 4) Số lượng sinh viên trong lớp có chiều cao lớn chiều cao TB của lớp.

```python
#{HỌC VIÊN VIẾT CODE VÀO ĐÂY}

import random

chieu_cao = tuple(round(random.uniform(1.5, 1.9), 2) for _ in range(12))
print(f"Tổng số sinh viên trong lớp: {len(chieu_cao)}")
print("Chiều cao của các sinh viên trong lớp :", chieu_cao)

cao_nhat = max(chieu_cao)
thap_nhat = min(chieu_cao)
print(f"1a. Sinh viên cao nhất: {cao_nhat} (m)")
print(f"1b. Sinh viên thấp nhất: {thap_nhat} (m)")

tb = sum(chieu_cao) / len(chieu_cao)
print(f"2. Chiều cao trung bình: {tb:.2f} m")

so_sv_bang_tb = chieu_cao.count(tb)
print(f"3. Số SV có chiều cao = trung bình: {so_sv_bang_tb}")

so_sv_lon_tb = sum(1 for h in chieu_cao if h > tb)
print(f"4. Số SV có chiều cao > trung bình: {so_sv_lon_tb}")

```

## Bài 16:
---
có 2 lớp sinh viên A1 và A2 mỗi lớp có 10 em, và bảng điểm của 2 lớp lưu ở biến danh sách như sau:
* A1 = [8, 7 , 9, 4, 5, 7, 9, 9, 8, 8]
* A2=  [10, 3 , 6, 10, 7, 7, 5, 3, 9, 0]

Yêu cầu:
1. Chuyển đổi A1, A2 về kiểu dữ liệu set
2. Liệt kê danh sách các đầu điểm mà 2 lớp nhận được
3. Các đầu điểm chung mà 2 lớp nhận được

```python
#{HỌC VIÊN VIẾT CODE VÀO ĐÂY}

A1 = [8, 7, 9, 4, 5, 7, 9, 9, 8, 8]
A2 = [10, 3, 6, 10, 7, 7, 5, 3, 9, 0]

set_A1 = set(A1)
set_A2 = set(A2)

print("Điểm lớp A1:", set_A1)
print("Điểm lớp A2:", set_A2)

tat_ca_diem = set_A1.union(set_A2)
print("Tất cả các đầu điểm mà 2 lớp nhận được:", tat_ca_diem)

diem_chung = set_A1.intersection(set_A2)
print("Các đầu điểm chung của 2 lớp:", diem_chung)






```

```python
#{HỌC VIÊN VIẾT CODE VÀO ĐÂY}

dict_point = {
    "Maths": 82,
    "Physics": 90,
    "History": 75,
    "Science": 94,
    "Biology": 75,
    "Chemistry": 88,
    "Information technology": 99
}

print("1.dict_point= ",dict_point)

subjects = list(dict_point.keys())
print("2. Danh sách tên môn học:", subjects)

max_score = max(dict_point.values())
min_score = min(dict_point.values())

subjects_max = [sub for sub, score in dict_point.items() if score == max_score]
subjects_min = [sub for sub, score in dict_point.items() if score == min_score]

print(f"3a. Điểm cao nhất của sinh viên: {max_score}\nMôn học: {subjects_max}")
print(f"3b. Điểm thấp nhất của sinh viên: {min_score}\nMôn học: {subjects_min}")





```

Nhập vào một số nguyên n, hãy viết chương trình sử dụng dictionary comprehension để tạo ra một dictionary trong đó keys là các số chẵn trong khoảng từ 0 --> n, values là bình phương của chính key đó (i, i^2)**

```python
#{HỌC VIÊN VIẾT CODE VÀO ĐÂY}

n = int(input("Nhập vào một số nguyên n: "))

dict_square = {i: i**2 for i in range(n + 1) if i % 2 == 0}

print(dict_square)





```
