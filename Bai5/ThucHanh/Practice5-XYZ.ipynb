**Yêu cầu: Tạo một vector gồm 30 phần tử, có giá trị tăng dần từ 1 đến 30**

Chỉ sử dụng các phương thức reshape, ravel, split…tách Vector a ở trên thành 3 vector con bao gồm:

1. a_lẻ: chứa các phần tử là số lẻ;
2. a_chan: chứa các phần tử là số chẵn.
3. a_3: chứa các phần tử chia hết cho 3

```python
import numpy as np
```

```python
#{HỌC VIÊN VIẾT CODE VÀO ĐÂY}

a = np.arange(1, 31)
print(a)
```

```python
a_le = a.ravel()[a.ravel() % 2 != 0]     # số lẻ
a_chan = a.ravel()[a.ravel() % 2 == 0]   # số chẵn
a_3 = a.ravel()[a.ravel() % 3 == 0]      # chia hết cho 3

print("a_lẻ:", a_le)
print("a_chẵn:", a_chan)
print("a_3:", a_3)
```

**Yêu cầu 3.1: Sử dụng vector_weigth trong bài thực hành 2 chuyển về ma trận kích thước 10 x 10**

```python
#{HỌC VIÊN VIÊT CODE VÀO ĐÂY}

v_weight_reshape = v_weight.reshape(10,10)

print(v_weight_reshape)
```

**Yêu cầu 3.2: Cho biết ma trận weight có tồn tại ma trân nghịch đảo không? Nếu có hãy xác định ma trận weight-1 ?**

```python
#{HỌC VIÊN VIÊT CODE VÀO ĐÂY}

det = np.linalg.det(v_weight_reshape)
print("Định thức =", det)
if det != 0:
    print("Ma trận có nghịch đảo.")
    weight_inv = np.linalg.inv(v_weight_reshape)
    print("Ma trận nghịch đảo weight⁻¹ =\n", weight_inv)
else:
    print("Ma trận không có nghịch đảo .")



```

**Yêu cầu 3.3: Tạo vector_diagonal chứa các phần tử trên đường chéo chính của ma trận weight, tính trace của ma trận weight.**

```python
#{HỌC VIÊN VIÊT CODE VÀO ĐÂY}

vector_diagonal = np.diagonal(v_weight_reshape)

trace_weight = np.trace(v_weight_reshape)

print("Vector đường chéo chính:", vector_diagonal)
print("Trace của ma trận weight:", trace_weight)

```

**Yêu cầu 3.4: Tìm giá trị lớn nhất, nhỏ nhất của ma trận đường chéo trên và ma trận đường chéo dưới không bao gồm các phần tử trên đường chéo chính**

```python
#{HỌC VIÊN VIÊT CODE VÀO ĐÂY}

vector_diagonal_below = np.diagonal(v_weight_reshape, offset=-1)

max_diagonal = np.max(vector_diagonal)
min_diagonal = np.min(vector_diagonal)

max_diagonal_below = np.max(vector_diagonal_below)
min_diagonal_below = np.min(vector_diagonal_below)

print("Đường chéo chính:", vector_diagonal)
print("Giá trị lớn nhất trên đường chéo chính:", max_diagonal)
print("Giá trị nhỏ nhất trên đường chéo chính:", min_diagonal)

print("Đường chéo dưới:", vector_diagonal_below)
print("Giá trị lớn nhất trên đường chéo dưới:", max_diagonal_below)
print("Giá trị nhỏ nhất trên đường chéo dưới:", min_diagonal_below)



```

## THỰC HÀNH SỐ 5
---
**Yêu cầu: Học viên tìm ma trận nghịch đảo và hạng của ma trận Height, Weight trong bài thực hành số 4:**


```python
#{HỌC VIÊN VIÊT CODE VÀO ĐÂY}

def check(matrix, name):
    if matrix.shape[0] == matrix.shape[1]:
      det = np.linalg.det(matrix)
      print(f"Định thức của ma trận {name}: {det}")
      if det != 0:
          inv_matrix = np.linalg.inv(matrix)
          print(f" Ma trận {name} có nghịch đảo:\n", inv_matrix)
      else:
          print(f" Ma trận {name} không có nghịch đảo (det = 0).")


check(height, "Height")
check(weight, "Weight")


```

```python
rank_height = np.linalg.matrix_rank(height)
rank_weight = np.linalg.matrix_rank(weight)

print("Hạng của ma trận Height:", rank_height)
print("Hạng của ma trận Weight:", rank_weight)
```
