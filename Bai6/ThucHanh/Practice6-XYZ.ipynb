4. Đặt lại tên các cột dữ liệu trong Dataframe:
    * Feature_1 --> Age
    * Feature_2 --> Gender
    * Feature_3 --> Type
    * Feature_4 --> Blood_pressure

```python
#{VIẾT CODE CỦA BẠN VÀO ĐÂY}

file_path = '/content/drive/MyDrive/for_student/Bai6/ThucHanh/Data_Excercise/Data_Patient.csv'

df_patient = pd.read_csv(file_path, index_col=0)

print("Thông tin tổng quan về tập dữ liệu:")
print(df_patient.info())

print("\n10 bệnh nhân đầu tiên:")
print(df_patient.head(10))

print("\n5 bệnh nhân cuối cùng:")
print(df_patient.tail(5))

df_patient.rename(columns={
    'feature_1': 'Age',
    'feature_2': 'Gender',
    'feature_3': 'Type',
    'feature_4': 'Blood_pressure',
    'feature_5': 'Cholesterol',
    'feature_6': 'Heartbeat',
    'feature_7': 'Thalassemia',
    'feature_8': 'Result'
}, inplace=True)

print("\n Sau khi đổi tên cột:")
print(df_patient.head())

print("\n Danh sách cột hiện tại:")
print(df_patient.columns.tolist())

```

**Yêu cầu 3.2:**
Sử dụng phương thức .describe() cho biết:
    
    * Thuộc tính Age:
        - Tuổi của bệnh nhân trẻ nhất
        - Tuổi của bệnh nhân già nhất
    * Thuộc tính Cholesterol:
        - Cholesterol trung bình của các bệnh nhân
        - Độ lệch chuẩn của giá trị này trong toàn bộ tập dữ liệu

```python
#{VIẾT CODE CỦA BẠN VÀO ĐÂY}

age_stats = df_patient['Age'].describe()
print("Thống kê thuộc tính Age:")
print(age_stats)

print(f"Tuổi nhỏ nhất: {age_stats['min']}")
print(f"Tuổi lớn nhất: {age_stats['max']}")


chol_stats = df_patient['Cholesterol'].describe()
print("\nThống kê thuộc tính Cholesterol:")
print(chol_stats)

print(f"Cholesterol trung bình: {chol_stats['mean']:.2f}")
print(f"Độ lệch chuẩn: {chol_stats['std']:.2f}")


male_count = (df_patient['Gender'] == 'Male').sum()
print(f"\nSố lượng bệnh nhân nam: {male_count}")


type_counts = df_patient['Type'].value_counts()
num_unique_type = df_patient['Type'].nunique()
most_common_type = type_counts.index[0]
most_common_count = type_counts.iloc[0]

print(f"\n Số lượng giá trị khác nhau của thuộc tính Type: {num_unique_type}")
print(f" Giá trị xuất hiện nhiều nhất: '{most_common_type}' ({most_common_count} lần)")


```

## THỰC HÀNH SỐ 5

Tập dữ liệu lưu trữ thông tin giao dịch tại cây Big Stree ATM (đặt tại một giao lộ sầm uất của Ấn Độ) theo thời gian từng ngày, cho biết Tổng số lần giao dịch và Số tiền giao dịch mỗi ngày (rupee - INR). Từ ngày 01-01-2014 đến 31-12-2019 (6 năm). Dữ liệu bao gồm 6 thuộc tính:
* Transaction Data: Thời gian
* ATM Name: Tên cây ATM
* No Of Withdrawals: Tổng số lần KH giao dịch tại ATM trong ngày
* Amount withdrawn Card: Số tiền giao dịch sử dụng thẻ của chính Ngân hàng
* Amount withdrawn Other Card: Số tiền giao dịch sử dụng thẻ của Ngân hàng khác
* Total amount Withdrawn: Tổng số tiền KH giao dịch tại ATM trong ngày

**YÊU CẦU:**
Sử dụng các kỹ thuật để Phân tích tập dữ liệu, tìm ra các thông tin có ích về thói quen, hành vi rút tiền tại cây ATM; giúp Ngân hàng đánh giá hiệu quả hoạt động của cây ATM và có kế hoạch tiếp quỹ hợp lý.


*(Xem lại bài phân ví dụ tích dữ liệu chuỗi thời gian người đi xe đạp!)*

```python
#{HỌC VIÊN VIẾT CODE VÀO ĐÂY}
path = '/content/drive/MyDrive/for_student/Bai6/ThucHanh/Data_Excercise/Big_Street_ATM.csv'
```

```python
df = pd.read_csv(path)

df['Transaction Date'] = pd.to_datetime(df['Transaction Date'])
df['Year'] = df['Transaction Date'].dt.year
df['Month'] = df['Transaction Date'].dt.month
df['DayOfWeek'] = df['Transaction Date'].dt.day_name()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.groupby('Year')['Total amount Withdrawn'].sum().plot(kind='bar')
plt.title("Tổng số tiền rút theo năm")
plt.ylabel("INR")
plt.show()

```

```python
df.groupby('Month')['Total amount Withdrawn'].sum().plot(kind='line')
plt.title("Tổng số tiền rút theo tháng (2014–2019)")
plt.xlabel("Month")
plt.ylabel("INR")
plt.grid(True)
plt.show()
```

```python
df.groupby('DayOfWeek')['No Of Withdrawals'].sum().reindex(
    ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
).plot(kind='bar')
plt.title("Lượng giao dịch theo ngày trong tuần")
plt.ylabel("Số giao dịch")
plt.show()
```

```python
df[['Amount withdrawn Card', 'Amount withdrawn Other Card']].sum()

df[['Amount withdrawn Card', 'Amount withdrawn Other Card']].sum().plot(kind='bar')
plt.title("Tổng tiền giao dịch theo loại thẻ")
plt.ylabel("INR")
plt.show()
```

```python
df.nlargest(10, 'Total amount Withdrawn')
```

```python
df_sorted = df.sort_values('Transaction Date')
df_sorted['MA_7'] = df_sorted['Total amount Withdrawn'].rolling(7).mean()

df_sorted[['Total amount Withdrawn','MA_7']].plot(figsize=(10,4))
plt.title("7-day Moving Average")
plt.show()
```
